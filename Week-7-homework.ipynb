```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('summons-15-25.csv')

# --- 1. Which borough had the most summons? ---
# We use value_counts() to count occurrences of each borough, and idxmax() to get the one with the highest count.
most_summons_boro = df['BORO'].value_counts().idxmax()
print(f"1. Borough with the most summons: {most_summons_boro}")
print("-" * 40)

# --- 2. How many times did the top 10 most common offenses occur? ---
# value_counts() sorts by frequency by default. We use head(10) to get the top 10.
top_10_offenses = df['OFFENSE_DESCRIPTION'].value_counts().head(10)
print("2. Top 10 most common offenses:")
print(top_10_offenses)
print("-" * 40)

# --- 3. What are the total number of summons broken down by year? ---
# We count the occurrences of each year and sort by the year index using sort_index()
summons_by_year = df['YEAR'].value_counts().sort_index()
print("3. Summons broken down by year:")
print(summons_by_year)
print("-" * 40)

# --- 4. What were some of the most unusual summons? ---
# "Unusual" is best defined here as the rarest or least frequently occurring summons. 
# We use tail(5) on the value_counts to find offenses that only happened 1 time.
unusual_summons = df['OFFENSE_DESCRIPTION'].value_counts().tail(5)
print("4. Most unusual summons (least common):")
print(unusual_summons)
print("-" * 40)

# --- 5. What is the count of summons by gender? ---
gender_counts = df['SEX'].value_counts()
print("5. Summons by gender:")
print(gender_counts)
print("-" * 40)

# --- 6. How many summons were issued by race? ---
race_counts = df['RACE'].value_counts()
print("6. Summons by race:")
print(race_counts)
print("-" * 40)

# --- 7. Which borough had the fewest offenses per 100K people? ---
# Using the exact population statistics provided
borough_populations = {
    'BROOKLYN': 2736074,
    'QUEENS': 2405464,
    'MANHATTAN': 1694251,
    'BRONX': 1472654,
    'STATEN ISLAND': 495747
}

# Get the count of summons per borough
boro_counts = df['BORO'].value_counts()

# Calculate the rate per 100,000 people for each borough
rates_per_100k = {}
for boro, count in boro_counts.items():
    if boro in borough_populations:
        # Formula: (Number of summons / Population) * 100,000
        rate = (count / borough_populations[boro]) * 100000
        rates_per_100k[boro] = rate

# Convert the dictionary to a Series and sort it from lowest to highest
rates_series = pd.Series(rates_per_100k).sort_values()
fewest_per_100k_boro = rates_series.idxmin()

print("7. Offenses per 100K people by borough (lowest to highest):")
print(rates_series)
print(f"\nBorough with fewest offenses per 100K: {fewest_per_100k_boro}")
print("-" * 40)
```
